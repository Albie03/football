```python
for pos in ['FWD', 'MID', 'DEF']:
    mask = df['position'] == pos

    Q1 = df.loc[mask, 'goals'].quantile(0.25)
    Q3 = df.loc[mask, 'goals'].quantile(0.75)
    IQR = Q3 - Q1

    lower = max(0, Q1 - 1.5 * IQR)
    upper = Q3 + 1.5 * IQR

    df.loc[mask, 'goals'] = df.loc[mask, 'goals'].clip(lower, upper)

# GK goals (special case)
df.loc[df['position'] == 'GK', 'goals'] = 0
```

```python
for pos in ['FWD', 'MID', 'DEF']:
    mask = df['position'] == pos

    Q1 = df.loc[mask, 'assists'].quantile(0.25)
    Q3 = df.loc[mask, 'assists'].quantile(0.75)
    IQR = Q3 - Q1

    lower = max(0, Q1 - 1.5 * IQR)
    upper = Q3 + 1.5 * IQR

    df.loc[mask, 'assists'] = df.loc[mask, 'assists'].clip(lower, upper)
```

```python
mask = df['position'] == 'GK'

Q1 = df.loc[mask, 'saves'].quantile(0.25)
Q3 = df.loc[mask, 'saves'].quantile(0.75)
IQR = Q3 - Q1

lower = max(0, Q1 - 1.5 * IQR)
upper = Q3 + 1.5 * IQR

df.loc[mask, 'saves'] = df.loc[mask, 'saves'].clip(lower, upper)

# Non-GK cannot have saves
df.loc[df['position'] != 'GK', 'saves'] = 0
```

```python
mask = df['position'].isin(['GK', 'DEF'])

Q1 = df.loc[mask, 'clean_sheets'].quantile(0.25)
Q3 = df.loc[mask, 'clean_sheets'].quantile(0.75)
IQR = Q3 - Q1

lower = max(0, Q1 - 1.5 * IQR)
upper = Q3 + 1.5 * IQR

df.loc[mask, 'clean_sheets'] = df.loc[mask, 'clean_sheets'].clip(lower, upper)

# MID/FWD cannot have clean sheets
df.loc[~mask, 'clean_sheets'] = 0
```

```python
df['minutes_played'] = df['minutes_played'].clip(upper=df['matches_played'] * 90,lower=0)

```

```python

df1=pd.get_dummies(df,columns=['position','league'],drop_first=True)
```

```python

df1= df1.drop(['player_name'],axis=1)
```

```python
x=df1.drop(['market_value_million_eur'],axis=1)
y=df1['market_value_million_eur']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_percentage_error
random=RandomForestRegressor(n_estimators=100,random_state=1)
random.fit(x_train,y_train)
random_pred=random.predict(x_test)
random_pred
```

```python
random_score=r2_score(y_test,random_pred)*100
random_score
```

```python
mape=mean_absolute_percentage_error(y_test,random_pred)*100
mape
```

```python
from xgboost import XGBRegressor
xgb=XGBRegressor(n_estimators=100,random_state=1)
xgb.fit(x_train,y_train)
xgb_pred=xgb.predict(x_test)
xgb_pred
```

```python
xgb_score=r2_score(y_test,xgb_pred)*100
xgb_score
```

```python
mape2=mean_absolute_percentage_error(y_test,xgb_pred)*100
mape2
```

```python
from sklearn.tree import DecisionTreeRegressor
tree=DecisionTreeRegressor()
tree.fit(x_train,y_train)
tree_pred=tree.predict(x_test)
tree_pred
```

```python
tree_score=r2_score(y_test,tree_pred)*100
tree_score
```

```python
mape3=mean_absolute_percentage_error(y_test,tree_pred)*100
mape3
```

```python
mod=pd.DataFrame({'algo':['RANDOM FOREST','XGBOOST','DECISION TREE'],'Score':[random_score,xgb_score,tree_score]})
mod.sort_values(by='Score',ascending=False)
```

```python
plt.figure(figsize=(9,6))
co=['green','red','blue']
bars=plt.bar(mod['algo'],mod['Score'],color=co)
plt.xlabel('MODEL')
plt.ylabel('SCORE')
plt.title('Model Performance Comparison')


for bar in bars:
    height=bar.get_height()
    plt.text(bar.get_x()+bar.get_width()/2,height,f'{height:.2f}',ha='center',va='bottom')
plt.tight_layout()
```
